```python
import yfinance as yf

# Question 1 - Extracting Tesla Stock Data Using yfinance
tesla = yf.Ticker("TSLA")
tesla_stock = tesla.history(period="max")
tesla_stock.reset_index(inplace=True)
tesla_stock.to_csv('Tesla_Stock.csv', index=False)
print(tesla_stock)


```

```python
# Question 2 - Extracting Tesla Revenue Data Using Webscraping

import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de Yahoo Finance
url = "https://finance.yahoo.com/quote/TSLA/history/?fr=sycsrp_catchall"

# Obtener el contenido de la página
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Crear un DataFrame vacío
tesla_data = pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"])

# Buscar las filas de la tabla
for row in soup.find("table").find("tbody").find_all("tr"):
    col = row.find_all("td")
    if len(col) >= 7:  # Verificar que haya al menos 7 columnas
        date = col[0].text.strip()
        Open = col[1].text.strip()
        high = col[2].text.strip()
        Low = col[3].text.strip()
        Close = col[4].text.strip()
        Adj_close = col[5].text.strip()
        Volume = col[6].text.strip()
        
        # Crear una fila temporal para el DataFrame
        new_row = pd.DataFrame({
            "Date": [date],
            "Open": [Open],
            "High": [high],
            "Low": [Low],
            "Close": [Close],
            "Adj Close": [Adj_close],
            "Volume": [Volume]
        })
        
        # Añadir la fila al DataFrame principal
        tesla_data = pd.concat([tesla_data, new_row], ignore_index=True)
    else:
        print(f"Incomplete row found: {col}")

# Mostrar los primeros registros del DataFrame
tesla_data.head(5)

```
